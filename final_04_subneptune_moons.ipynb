```python
# import the planets from Hadden & Lithwick (2017)
hadden = import_hadden()
# take only the planets marked as "robust"
hadden = hadden[hadden["Robust"] == 1]

# import the planets from the Berger et al. (2020) exoplanet catalog
berger = import_berger2020a()
# turn the abbreviated KOI into a full KOI
berger["KOI_full"] = get_full_KOI_berger(berger["KOI"])

# import a .csv file from the exoplanet archive
archive = import_exoplanet_archive_KOIs()
# match the planets from Berger et al. (2020) to some additional
# information from the exoplanet archive
df2 = berger.merge(archive, left_on = "KOI_full", right_on = "kepoi_name")
```

```python
# calculate the semimajor axes of the TTV planets
hadden["a"] = (hadden["P"] / 365.25)**(2/3) * (hadden["M_st"])**(1/3)
```

```python
# define an expression from Barnes & O'Brien (2002)
# for the maximum-mass prograde moon that could survive
# for a 4.5 Gyr system history

R_UR = 3.981 # radius of uranus
R_SAT = 9.51 # radius of saturn

def max_moon(planet_mass, star_mass, semimajor_axis, planet_radius, planet_Q):
    # planet_mass -- earth masses
    # star_mass -- solar_masses
    # semimajor_axis -- au
    # planet_radius -- earth radii
    
    # we've already assumed Q to be 10^3 in this scaling relation
    # so planet_Q = 100 corresponds to tidal Q = 10^5 (gas giant)
    
    # frame the equation as a scaling relation    
    largest_satellite = 4.07E-6 *\
        planet_Q *\
        (semimajor_axis/0.1)**(39/6) *\
        (star_mass)**(-13/6) *\
        (planet_mass)**(8/3) *\
        (planet_radius)**(-5)
    
    return largest_satellite / planet_mass
```

```python
# define the Fulton Gap
def fulton(P, m=-0.09, a=0.37):
    return 10**(m * np.log10(P) + a)

# define the cutoff between super-Earth/sub-Neptune
def super_Earth(P, R, m=-0.09, a=0.37):
    # find the maximum allowable radius
    # for a super-Earth
    maximum_SE = fulton(P, m=m, a=a)
    # compare the radius of the planet
    # to the maximum allowable radius
    return R < maximum_SE

# assign Q values to the columns
# largest satellite is linear with tidal Q
# determine tidal Q

tidal_Q_arr = []
for i in range(len(hadden)):
    p = hadden.iloc[i]
    if super_Earth(p["P"], p["R_pl"]):
        tidal_Q_arr.append(0.1) # x 10^3
    else:
        if p["R_pl"] < R_UR:
            tidal_Q_arr.append(1) # x 10^3 (actually corresponds to Q = 10^3)
        elif p["R_pl"] > R_SAT:
            tidal_Q_arr.append(100) # x 10^3 (actually corresponds to Q = 10^5)
        else:
            tidal_Q_arr.append(10) # x 10^3 (actually corresponds to Q = 10^4)

hadden["Q"] = np.array(tidal_Q_arr)

# get the maximum moon size
hadden["maxmoon"] = max_moon(hadden["M_pl"], hadden["M_st"], hadden["a"], hadden["R_pl"], hadden["Q"])

CUTOFF = 10 # earth masses
# divide the Hadden & Lithwick (2017) TTV planets into
# those that do/don't possibly host pea-like exomoons
peamoons = []
for i in range(len(hadden)):
    p = hadden.iloc[i]
    if p["maxmoon"] <= 1.0E-5:
        peamoons.append(0)
    else:
        if p["M_pl"] > CUTOFF:
            peamoons.append(1)
        else:
            peamoons.append(2)

hadden["peamoons"] = np.array(peamoons)

hosts = hadden[hadden["peamoons"] == 1]
possiblehosts = hadden[hadden["peamoons"] == 2]
nohost = hadden[hadden["peamoons"] == 0]
```

```python
# generate the figure

fig, ax = plt.subplots()
fig.set_figwidth(5)
fig.set_figheight(5)

# plot the classic radius versus period diagram
ax.scatter(df2["koi_period"], df2["Radius"], s = 1, c = "grey", alpha = 0.3)

# plot these planets on the P-R diagram
ax.scatter(hosts["P"], hosts["R_pl"], s = 50, label = "Regular Moons Predicted", marker = "o", c = "olivedrab")
ax.scatter(possiblehosts["P"], possiblehosts["R_pl"], s = 30, marker = "x", c = "firebrick")
ax.scatter(nohost["P"], nohost["R_pl"], s = 30, label = "Regular Moons Impossible", marker = "x", c = "firebrick")

# make the axes log-scale
ax.set_xscale("log")
ax.set_yscale("log")

# set the axis bounds
ax.set_xlim((2, 200))
ax.set_ylim((0.4, 15))

# plot the Fulton gap
periods_fulton = np.linspace(3, 150, 100)
ax.plot(periods_fulton, fulton(periods_fulton), c = "black", lw = 1, ls = "--")

# plot a legend
ax.legend(fontsize = 11, framealpha = 1)

# label the axes
ax.set_xlabel("Host Planet Period [d]", fontsize = 14)
ax.set_ylabel("Host Planet Radius [R$_{\oplus}$]", fontsize = 14)

# resize the axis tick markers
ax.tick_params(axis='both', which='major', labelsize=12)

# set the y-axis and x-axis ticks
ax.set_yticks([0.5, 1, 3, 5, 10, 15])
ax.set_yticklabels(["0.5", "1.0", "3.0", "5.0", "10.0", "15.0"])
ax.set_xticks([3, 5, 10, 30, 60, 100])
ax.set_xticklabels(["3", "5", "10", "30", "60", "100"])

# add a label for the Fulton gap
ta = ax.text(45, 1.0, "Fulton Gap", fontsize = 12)
ta.set_bbox(dict(facecolor='white', alpha=1, edgecolor='none'))

# add an arrow for the Fulton gap label
ax.plot(np.array([55, 55]), np.array([1.05, 1.4]), c = "black", lw = 3)
ax.scatter(55, 1.45, marker = "^", s = 50, c = "black", zorder = 10)

ax.tick_params(axis="y",direction="in", which="both")
ax.tick_params(axis="x",direction="in", which="both")

ax.tick_params(bottom=True, top = True, left = True, right = True, which = "both")

# make the formatting nice
fig.tight_layout()

saveFigure = False
if saveFigure: plt.savefig("satellitepeas.png", dpi = 800)
```
